# Solución de un PVI segundo orden lineal homogéneo de coeficientes constantes

Solucionemos la ED

$$ y'' - y = 0 \ \ \ \ \:\:\:\:\:  y=y'' $$


# Clasificación de la ED dada

## Solución de la ED dada

$$ y = c_{1}e^{x} + c_{2}e^{-x} $$

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.linspace(-3,3, 100)
C1 = [0,1]
C2 = [0,1]

for c1 in C1:
    for c2 in C2:
        y = c1*np.exp(x) + c2*np.exp(-x)
        plt.plot(x,y)

plt.title('Algunas soluciones de PVI de segundo orden')        
plt.savefig('solu_pvi_1')        
plt.show()        
```

# Grafica de ecuaciones implícitas con Matplotlib

```python
import matplotlib.pyplot
from numpy import arange
from numpy import meshgrid

delta = 0.025
xrange = arange(-5.0, 5.0, delta)
yrange = arange(-5.0, 5.0, delta)
X, Y = meshgrid(xrange,yrange)
C = [-1,0,1]
for c in C:
    F = Y*X
    G = c
    matplotlib.pyplot.contour(X, Y, (F - G))
    
    
matplotlib.pyplot.show()
```

# Graficación de una solución dada implícitamente

$$ ydx + xdy = 0  $$

$$ M = y, \ \ \ \ \ \  \ \ N = x $$

$$My = Nx $$

Como $M_{y} = 1 = N_{x}$, entonces la ED dada es exacta y por tanto el lado izquierdo de esta ecuación debe la diferencial de alguna función $f(x,y)$. 

$$ f_{x} = M = y$$

$$ f = \int Mdx = \int ydx = xy + g(y) $$

$$ f_{y} = N $$

$$ x + g'(y) = x $$

$$ g'(y) = 0 $$

$$ g(y) = c $$

$$ xy = c $$

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.arange(-3,4,0.1)
y = np.arange(-3,4,0.1)
X,Y = np.meshgrid(x,y)

F=X*Y
G = -1

plt.contour(X,Y,F-G)
plt.grid() 
```

## Observaciones o inferencias que del grafico se pueden obtener

1. cada una de las soluciones de la ED es decreciente a cero.
2. Cada una de las soluciones es una hipérbola.
3. Cuando $c = 0$, entonces la solución de la ED es dada por la ecuación $xy = 0$.
4. Hay soluciones de la ED que son funciones crecientes y estas se obtienen para los valores de $c$ negativos.
