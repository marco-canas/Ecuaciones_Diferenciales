#  PVI LINEALES DE SEGUNDO ORDEN

* Resolver la ecuacion lineal de segundo orden y graficar. 

$$4y^{''\:}-8y^{'\:}+3y=0$$ 

$$y\left(0\right)=2$$ 

$$y\left(0\right)=\frac{1}{2}$$

* Supongo que $$y=e^{mx}$$ es solucion, es decir, supongo que la solución de este problema es de forma exponencial.  

$$y'=me^{mx}\:\:\:\:y''=m^2e^{mx}$$

$$4\left(m^2e^{mx}\right)-8\left(me^{mx}\right)+3\left(e^{mx}\right)=0$$

$$4m^2e^{mx}-8me^{mx}+3e^{mx}=0$$

$$e^{mx}\left(4m^2-8m+3\right)=0$$

* Utilizo la formula general para hallar los **m**

$$m=-\frac{b+-\sqrt{b^2-4ac}}{2a}$$

$$m=\frac{8+-\sqrt{8^2-4\left(4\right)\left(3\right)}}{2\left(4\right)}$$

$$m=\frac{8+-\sqrt{64-48}}{8}$$

$$m_1=\frac{8+4}{8}\:\:\:\:\:\:\:\:\:m_2=\frac{8-4}{8}$$

$$y=e^{\frac{3}{2\:}x}\:\:y=e^{\frac{1}{2\:}x}\:$$

$$y=C_1e^{\frac{3}{2\:}x}\:\:y=C_2e^{\frac{1}{2\:}x}\:\:$$

$$4y^{''\:}-8y^{'\:}+3y=0$$

$$\:\:\:y\left(0\right)=2\:\:\:\:y'\left(0\right)=\frac{1}{2}$$

* La solucion general es 

$$y=C_1e^{\frac{3}{2}x}+\:C_2e^{\frac{1}{2}x}$$

entonces la solución del PVI es $y'=\frac{3}{2}C_1e^{\frac{3}{2}x}+\:\frac{1}{2}C_2e^{\frac{1}{2}x}$

* ahora reemplazamos las condiciones iniciales

$$\frac{1}{2}=\frac{3}{2}C_1e^{\frac{3}{2}\left(0\right)}+\frac{1}{2}C_2e^{\frac{1}{2}\left(0\right)}$$

$$2=C_1e^{\frac{3}{2}\left(0\right)}+C_2e^{\frac{1}{2}\left(0\right)}$$

* Realizamos sistema de ecuacion $2\cdot 2$

$$\frac{1}{2}=\frac{3}{2}C_1+\frac{1}{2}C_2$$

$$2=C_1+C_2$$

* Multiplicamos la ecuacion 1 por $-\frac{2}{3}$ y tenemos:

$$-\frac{1}{3}=-C_1-\frac{1}{3}C_2$$

$$2=C_1+C_2$$

* Cancelamos $C_1$ para hallar el valor de $C_2$

$$\frac{5}{3}=\frac{2}{3}C_2$$

$$C_2=\frac{5}{3}\cdot \frac{3}{2}=\frac{5}{2}$$

* Remplazamos el valor de $C_2$ en la segunda ecuación para hallar $C_1$

$$2=C_1+\frac{5}{2}$$

$$2-\frac{5}{2}=C_1$$

$$C_1=-\frac{1}{2}$$

* Reemplazamos $C_1$ y $C_2$ en $Y$

$$y=-\frac{1}{2}e^{\frac{3}{2}x}+\frac{5}{2}e^{\frac{1}{2}x}→\:solucion\:explicita\:del\:PVI\:dado$$
