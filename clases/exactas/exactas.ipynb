## Definici칩n de ecuaci칩n exacta 

Una expresi칩n diferencial $M(x, y)dx + N(x, y)dy$ es una diferencial exacta en una regi칩n $R$ del plano $xy$ si 칠sta corresponde a la diferencial de alguna funci칩n $f(x, y)$ definida en $R$.   

Una ecuaci칩n diferencial de primer orden de la forma

$$ M(x,y)dx + N(x,y)dy = 0 $$

se dice que es una **ecuaci칩n exacta** si la expresi칩n del lado izquierdo es una diferencial exacta. 


## Teorema: Criterio para una diferencial exacta  

Sean $M(x, y)$ y $N(x, y)$ continuas y que tienen primeras derivadas parciales continuas en una regi칩n rectangular $R$ definida por $a < x < b$, $c<y<d$.  

Entonces una condici칩n necesaria y suficiente para que $M(x, y) dx + N(x, y) dy$ sea una 
diferencial exacta es 

$$ \frac{\partial M}{\partial y} = \frac{\partial N}{\partial x} $$
