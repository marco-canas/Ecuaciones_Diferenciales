## Ecuaciones diferenciales de segundo orden: 




La solución de una ecuación diferencial de la forma $y'' = f(x,y,y')$ suele comprender dos constantes arbitrarias. A la inversa, es posible demostrar que una familia dada de funciones que contiene dos constantes arbitrarias es solución de una ecuación diferencial de segundo orden. 

Elimine las constantes $c_{1}$ y $c_{2}$ utilizando a $y$, $y'$ y $y''$, para encontrar la ecuación diferencial que satisface la familia dada de funciones: 

$$ y = c_{1}\cos(x) + c_{2}\sin(x) $$
Página 144.  
Respuestas en la página 711. 

## Solución: 

$$ y = c_{1}\cos(x) + c_{2}\sin(x) \quad \quad \Rightarrow \quad \quad y' = -c_{1}\sin(x) + c_{2}\cos(x) $$

$$ y'' = -c_{1}\cos(x) - c_{2}\sin(x) = -(c_{1}\cos(c) + c_{2}\sin(x)) = - y $$

$$ y'' = -y \quad \quad \iff \quad \quad  y'' + y = 0 $$

## Verificación analítica: 

$y'' - y = 0$ es una ecuación diferencial de segundo orden lineal homogénea de coeficientes constantes, de donde se infiere que tiene soluciones de la forma $y = e^{rx}$. 

## Verificación con sympy

```python
from sympy import * 
x,y = symbols('x y')
y = Function('y')(x)
c1,c2 = symbols('c1 c2') 
```

### Verificación resolviendo la ED obtenida

```python
ed = diff(y,x,2) + y 
dsolve(ed,y) 
```

## Gráfica de algunas de las funciones de la familia: $y = c_{1}\cos(x) + c_{2}\sin(x)$

```python
# y = 2(x+c)^(-2) 
import numpy as np 
import matplotlib.pyplot as plt 
plt.style.use('dark_background')
from ipywidgets import interact 

C = [0,1]
fig = plt.figure()
ax = fig.add_subplot()

x = np.linspace(0,10, 100)

for c1 in C:
    for c2 in C: 
        y = c1*np.cos(x) + c2*np.sin(x) 
        ax.plot(x,y) 
    
ax.grid(True, alpha = 0.3) 
plt.savefig('ejer_34_curvas.jpg')   

plt.show()
```
