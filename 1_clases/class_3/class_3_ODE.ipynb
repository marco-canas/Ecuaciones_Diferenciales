## Ejemplo 1: Decaimiento radiactivo  

Página 61 del Boyce DiPrima 

El isótopo radiactivo torio 234 se desintegra con una rapidez proporcional a la cantidad presente del mismo.

Si $100$ mg de este material se reducen a $82.04$ mg en una semana, encontrar una expresión para la cantidad presente en cualquier instante.  

También halle el intervalo de tiempo que debe transcorrir para que la masa decaiga hasta la mitad de su valor inicial. 

## Solución:

Sea $Q(t)$ la cantidad de torio 234 presente en cualquier instante $t$,
* en donde $Q$ se mide en miligramos y
* $t$ en días. 

La observación física de que el torio 234 se desintegra con rapidez proporcional a la cantidad presente significa que la razón de cambio con el tiempo $dQ/dt$ es proporcional a $Q$, por tanto $Q$ satisface la ecuación diferencial 

$$ \frac{dQ}{dt} = -rQ $$

en donde la constante $r > 0$ se conoce como **la razón de decaimiento**.

Se busca pa solución de la ecuación anterior que también satisfaga la condición inicial 

$$ Q(0) = 100 $$

así como la condición 

$$ Q(7) = 82.04 $$


La ecuación 

$$ \frac{dQ}{dt} = -rQ $$

es lineal y también separable. 


Su solución general es

$$Q(t) = ce^{-rt} $$

donde $c$ es una constante arbitraria.

La condición $$ Q(0) = 100 $$ requiere que $c = 100$.

Y por lo tanto, 

$$Q(t) = 100e^{-rt} $$


A fin de satisfacer la condición $Q(7) = 82.04$, se hace $t = 7$ y $Q = 82.04$ en la ecuación anterior.   

Con ello obtenemos 

$$ 82.04 = 100e^{-7r} $$


Entonces 

$$r = - \frac{\ln 0.8204}{7} = 0.02828 \ \text{días}^{-1} $$

Por tanto, se ha determinado la razón de decaimiento $r$. 

Si se usa este valor de $r$ en 

$$Q(t) = 100e^{-rt} $$

se obtiene 

$$Q(t) = 100e^{-0.02828t}\ mg. $$

con lo que se obtiene el valor de $Q(t)$ en cualquier instante. 


El lapso durante el que la masa se reduce a la mitad de su valor original se denomina **vida media** del material. 

Si $\tau$ es el tiempo en que $Q(t)$ es igual a $50$ mg, entonces 

$$ 50 = 100e^{rt} $$


o bien 

$$r\tau = \ln 2$$


La relación anterior que expresa el vínculo entre la razón de decaimiento y la vida media es válida no solo para el Torio 234 sino para cualquier material que obedezca la ecuación diferencial 

$$ \frac{dQ}{dt} = -rQ(t) $$


al usar el valor de $r = 0.02828$ se encuantra que para el torio 234 

$$\tau = \frac{\ln 2}{0.02828} \approx 24.5 \text{días} $$


## Ejemplo 2 Interés compuesto
Página 62 del Boyce DiPrima 

Supóngase que, en un banco, se deposita una cantidad de dinero o fondo $S(t_{0})$, que paga interés a una tasa anual $r$. 

El valor $S(t)$ de la inversión en cualquier instante $t$ depende de la frecuencia con la que se componga el interés, así como de la tasa de éste. 

Las instituciones financieras tienen varias políticas sobre la composición:
* en algunas se compone mensualmente  
* en otras semanalmente,  
* y en otras incluso diariamente. 

Si se supone que la composición se llevó a cabo *continuamente*, entonces es posible plantear un problema sencillo con valor inicial que describa el crecimiento de la inversión. 

La razón de cambio del valor de la inversión es $dS/dt$, y esta cantidad es igual a la rapidez con que se acumula el interés, que es la tasa de interés $r$ multiplicada por el valor actual de la inversión $S(t)$. 

Por tanto, 

$$ \frac{dS}{dt} = rS(t) $$

es la ecuación diferencial que rige el proceso. 

Supóngase que también se conoce el valor de la inversión en algún instante particular, por ejemplo

$$S(0) = S_{0} $$


Entonces, la solución del problema con valor inicial 

\begin{align*}
\frac{dS}{dt} & = rS(t),\\
S(0) & = S_{0}
\end{align*}
da el balance $S(t)$ de la cuenta en cualquier instante $t$.

Este problema con valor inicial se resuelve con facilidad. 

De hecho, es el mismo, salvo por el signo de la constante de proporcionalidad, que el problema con valor inicial anterior que describe el decaimiento radiactivo. 

Como consecuencia, al resolver las ecuaciones anteriores se obtiene que 

$$ S(t) = S_{0}e^{rt} $$



Por tanto, una cuenta bancaria con interés compuesto en forma continua crece exponencialmente. 

Comparese ahora los resultados del modelo continuo que acaba de describirse con la situación en la que la composición ocurre en intervalos de tiempo finitos. 

Si el interés se compone una vez al año, entonces al cabo de $t$ años 

$$ S(t) = S_{0}(1+r)^{t} $$


Si el interés se compone 2 veces al año, entonces al término de seis meses el valor de la inversión es 

$$ S_{0}\left( 1+\frac{r}{2} \right), $$


y al cabo de un año es 

$$S_{0}\left( 1 + \frac{r}{2} \right)^{2}$$


Entonces, después de $t$ años se tiene 

$$S(t) = S_{0}\left( 1 + \frac{r}{2} \right)^{2t}$$


En general, si el interés se compone $m$ veces al año, entonces 

$$S(t) = S_{0}\left( 1 + \frac{r}{m} \right)^{mt}$$


La relación entre las fórmulas 

$$ S(t) = S_{0}e^{rt} $$

y 

$$S(t) = S_{0}\left( 1 + \frac{r}{m} \right)^{mt}$$

se aclara si se recuerda lo visto en Cálculo que 

$$ \lim_{m \to \infty} S_{0}\left( 1 + \frac{r}{m} \right)^{mt} = S_{0}e^{rt} $$

Para tomar en cuenta el hecho de que la tasa de crecimiento en realidad depende de la población, en la ecuación 

$$ \frac{dN}{dt} = rN, $$

se reemplaza la constante $r$ por la función $f(N)$, con lo que se tiene la ecuación modificada

$$ \frac{dN}{dt} = f(N)N $$

Ahora se debe elegir $f(N)$ de modo que $f(N) \cong r > 0 $ cuando $N$ sea pequeño, $f(N)$ decrezca cuando $N$ aumenta, y $f(N) < 0$ cuando $N$ sea suficientemente grande. 

La función más sencilla que tiene estas propiedades es $f(N) = r - aN$, donde $a$ es también una constante positiva. 

Si se usa esta función en  

$$ \frac{dN}{dt} = f(N)N $$

obtenemos 

$$\frac{dN}{dt} = (r-aN)N $$

Esta ecuación se conoce como la ecuación de Verhulst o ecuación logística. 

A menudo resulta conveniente escribir la ecuación logística en la forma equivalente 

$$ \frac{dN}{dt} =  r\left(1 - \frac{N}{K} \right)N $$


donde $k = r/a$. 

La constante $r$ se denomina **razón de crecimiento intrínseco**, es decir, la razón de crecimiento en ausencia de todo factor limitante.  

La solución de la ecuación logística 


$$ \frac{dN}{dt} =  r\left(1 - \frac{N}{K} \right)N $$

puede encontrarse fácilmente por el método de separación de variables. 

Ahora se mostrará que al aplicar el razonamiento geométrico es posible describir directamente las características principales de la solución a partir de la propia ecuación diferencial sin resolverla. 

Esto es importante porque amenudo pueden aplicarse los mismos métodos en ecuaciones más complicadas, cuyas soluciones son más difíciles de obtener. 

En la figura 2.6.2 se muestra la gráfica de $dN/dt$ contra $N$, en donde $dN/dt$ está dada por el segundo miembro de la ecuación 

$$ \frac{dN}{dt} =  r\left(1 - \frac{N}{K} \right)N $$

La gráfica es una parábola que se interseca con el eje $N$ en $(0,0)$ y en $(k,0)$ y con vértice $(k/2,rk/4)$. 

Para $0 < N < k$, se ve que $dN/dt > 0$ y, por lo tanto $N$ es función creciente de $t$, esto se indica con la flecha que indica hacia la derecha cerca del eje $N$. 

De manera semejante, si $N$ le gana a $K$, entonces $dN/dt < 0$; de donde, $N(t)$ es decreciente, como se indica por la flecha que apunta hacia la izquierda.  

Si $N = 0$ o $N = K$, entonces $dN/dt = 0$ y $N(t)$ no cambia. 

Las soluciones constante $\phi_{1}(t) = 0$ y $\phi_{2}(t) = k$ **soluciones de equilibrio**, a las cuales corresponden los puntos $N = 0$ y $N=K$, del eje $N$ se denominan **puntos de equilibrio** o **puntos críticos**.
